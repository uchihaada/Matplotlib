```python
import matplotlib.pyplot as plt

# define the list of 2D points
points = [(1, 2), (3, 5), (6, 3), (8, 7)]

# extract the x and y values from the points
x_values = [p[0] for p in points]
y_values = [p[1] for p in points]

# set the figure size and axis limits
fig, ax = plt.subplots(figsize=(8, 6))
ax.set_xlim(min(x_values) - 4, max(x_values) + 4)
ax.set_ylim(min(y_values) - 4, max(y_values) + 4)

# plot the points with red color and marker style '+'
ax.plot(x_values, y_values, linestyle='--', linewidth=2.5, color='red', marker='+', markersize=10.5)


ax.set_title('2D Points Plot')
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')


plt.savefig('r1.jpg')
```

```python
import matplotlib.pyplot as plt
import numpy as np

# define the x values
x_values = np.linspace(-5, 5, 100)

# define the y values for the functions
y_values_1 = x_values ** 2
y_values_2 = x_values ** 3

# set the figure size and axis limits
fig, ax = plt.subplots(figsize=(8, 6))
ax.set_xlim(-5, 5)
ax.set_ylim(-30, 30)

# plot the functions with different colors and labels
ax.plot(x_values, y_values_1, label='f(x)=x^2', color='red')
ax.plot(x_values, y_values_2, label='f(x)=x^3', color='blue')

# add legend and axis labels
ax.legend()
ax.set_title('Functions Plot')
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')

# save the figure as 'functions.jpg'
plt.savefig('functions.jpg')
```

```python
import matplotlib.pyplot as plt
import numpy as np

# define the x values
x_values = np.linspace(-1, 1, 100)

# define the y values for the functions
y_values_1 = x_values
y_values_2 = x_values ** 2
y_values_3 = x_values ** 3
y_values_4 = x_values ** 4
y_values_5 = x_values ** 5
y_values_6 = x_values ** 6

# create a figure with 6 subplots
fig, axs = plt.subplots(2, 3, figsize=(12, 8))

# plot each function in a different subplot
axs[0, 0].plot(x_values, y_values_1)
axs[0, 0].set_title('f(x) = x')
axs[0, 0].set_xlabel('x')
axs[0, 0].set_ylabel('y')

axs[0, 1].plot(x_values, y_values_2)
axs[0, 1].set_title('f(x) = x^2')
axs[0, 1].set_xlabel('x')
axs[0, 1].set_ylabel('y')

axs[0, 2].plot(x_values, y_values_3)
axs[0, 2].set_title('f(x) = x^3')
axs[0, 2].set_xlabel('x')
axs[0, 2].set_ylabel('y')

axs[1, 0].plot(x_values, y_values_4)
axs[1, 0].set_title('f(x) = x^4')
axs[1, 0].set_xlabel('x')
axs[1, 0].set_ylabel('y')

axs[1, 1].plot(x_values, y_values_5)
axs[1, 1].set_title('f(x) = x^5')
axs[1, 1].set_xlabel('x')
axs[1, 1].set_ylabel('y')

axs[1, 2].plot(x_values, y_values_6)
axs[1, 2].set_title('f(x) = x^6')
axs[1, 2].set_xlabel('x')
axs[1, 2].set_ylabel('y')

# set the overall title and adjust the spacing between subplots
fig.suptitle('Functions Plot')
fig.subplots_adjust(hspace=0.4, wspace=0.3)

# save the figure as 'functions.jpg'
plt.savefig('functions.jpg')
```

```python
import matplotlib.pyplot as plt

def plot_functions(x):
    # Define the two functions
    f1 = lambda x: 2*x + 4
    f2 = lambda x: 4*x**2 + 5*x + 4

    # Evaluate the functions for the given x values
    y1 = [f1(xi) for xi in x]
    y2 = [f2(xi) for xi in x]

    # Plot the functions
    plt.plot(x, y1, label="f(x) = 2x + 4")
    plt.plot(x, y2, label="f(x) = 4x^2 + 5x + 4")

    # Add labels and legend
    plt.xlabel("x")
    plt.ylabel("y")
    plt.title("Two Functions")
    plt.legend()

    # Save and show the plot
    plt.savefig("functions.png")
    plt.show()
x = eval(input())
plot_functions(x)
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Define the data
data = [4,5,5,7,6,8,8,9,4,4,4,4,5,5,7,7,8,8,8]

# Define the histogram parameters
bin_width = 5
bins = np.arange(min(data), max(data) + bin_width, bin_width)

# Plot the histogram
plt.hist(data, bins=bins, edgecolor='black')
plt.title('Plotting Histogram')
plt.xlabel('Values')
plt.ylabel('Frequency')

# Set the x-axis limits
plt.xlim(min(data)-1, max(data)+5)

# Save the plot and show it
plt.savefig('histogram.png')
plt.show()
```

```python
#@title Default title text
import matplotlib.pyplot as plt

# Define the data
months = range(1, 13)
faceCreamSalesData = [10, 12, 22, 23, 17, 18, 21, 24, 19, 13, 16, 18]
faceWashSalesData = [12, 15, 20, 22, 20, 18, 15, 16, 14, 15, 13, 10]
toothPasteSalesData = [18, 21, 17, 19, 23, 24, 20, 22, 19, 15, 13, 16]
bathingsoapSalesData = [15, 18, 16, 14, 19, 21, 17, 20, 19, 13, 11, 14]
shampooSalesData = [20, 22, 17, 21, 15, 16, 23, 18, 20, 14, 15, 12]
moisturizerSalesData = [7, 10, 13, 15, 12, 11, 8, 10, 9, 12, 11, 7]

# Plot the multiline plots
plt.plot(months, faceCreamSalesData, label='Face Cream')
plt.plot(months, faceWashSalesData, label='Face Wash')
plt.plot(months, toothPasteSalesData, label='Toothpaste')
plt.plot(months, bathingsoapSalesData, label='Bathing Soap')
plt.plot(months, shampooSalesData, label='Shampoo')
plt.plot(months, moisturizerSalesData, label='Moisturizer')

# Set the title of the plot
plt.title('Product Sales')

# Set the labels for the x and y axes
plt.xlabel('Month')
plt.ylabel('Units Sold')

# Set the tick marks for the x and y axes
plt.xticks(months)
plt.yticks(range(0, 30, 5))

# Set the location of the legend
plt.legend(loc='upper left')

# Show the plot
plt.show()
```
